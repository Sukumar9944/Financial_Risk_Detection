---

##### **Removing negative sign from DAYS_BIRTH, DAYS_EMPLOYED, DAYS_REGISTRATION, DAYS_ID_PUBLISH DAYS_LAST_PHONE_CHANGE columns**

```python
def remove_negative(num):
    num = abs(num)
    return num
```

```python
new_df['DAYS_BIRTH'] = new_df['DAYS_BIRTH'].apply(remove_negative)
```

```python
new_df['DAYS_EMPLOYED'] = new_df['DAYS_EMPLOYED'].apply(remove_negative)
```

```python
new_df['DAYS_REGISTRATION'] = new_df['DAYS_REGISTRATION'].apply(remove_negative)
```

```python
new_df['DAYS_ID_PUBLISH'] = new_df['DAYS_ID_PUBLISH'].apply(remove_negative)
```

```python
new_df['DAYS_LAST_PHONE_CHANGE'] = new_df['DAYS_LAST_PHONE_CHANGE'].apply(remove_negative)
```

---

##### **Removing Unknowns from Gender Column**

```python
new_df = new_df.query("CODE_GENDER != 'XNA'")
```
